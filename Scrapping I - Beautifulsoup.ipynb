## **BeautifulSoup**

### Import Module

```python
# Import necessary libraries

import requests #Sending and receiving HTTP requests
from bs4 import BeautifulSoup #Parsing and navigation of HTML and XML web pages
import pandas as pd #Used for data manipulation and analysis
```

### Get Response from URL

```python
# URL for the webpage to be scraped
url = 'https://www.bbc.com/sport/football/premier-league/table'

# Send a request to the webpage
response = requests.get(url)

# Print response
print(response)
```

### Parse & Print HTML with BeautifulSoup


```python
# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')

print(soup.prettify())
```

### Get Table Data from BeautifoulSoup

```python
# Get Div with Class table-2
table = soup.find('table', {'class': 'ssrcss-5o3au1-Table e13j9mpy3'})

print(table.prettify())
```

### Extract Table Headers and Rows Data

```python
# Understanding How to get table headers

x = table.find_all('th')
for i, y in enumerate(x):
  print(i, y)
  print(i, "text:", y.text)
# print(table)
```

```python
# Extract the table headers
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

print(headers) #table header ke-10 berbeda namanya dengan web yang di scrapping, oleh karena itu nama headernya akan dirubah
```

```python
headers = headers[0:9]
headers.append('Form')

print(headers)
```

```python
# Understanding How to get table headers

x = table.find_all('tr')
for i, y in enumerate(x):
  print(i, y.prettify())
  print(i, "text:", y.text)
# print(table)
```

```python
# Extract the table data
data = []
for tr in table.find_all('tr'):
    row_data = []
    for td in tr.find_all('td'):
        row_data.append(td.text.strip())
    if len(row_data) > 0:
        data.append(row_data)
```

```python
data
```

```python
# Tampilan data kolom pertama hasil scraping berbeda dengan tampilan pada web, sehingga diperlukan sedikit modikasi untuk menyesuaikannya
# Kolom pertama

import re

for row in data:
    # Ambil nama tim di kolom pertama (misal '1Arsenal' atau '10Crystal Palace')
    nama_tim = row[0]

    # Hapus angka di awal menggunakan regex
    nama_tim_bersih = re.sub(r'^\d+', '', nama_tim).strip()

    # Simpan kembali ke list
    row[0] = nama_tim_bersih
```

```python
# Tampilan data kolom terakhir hasil scraping berbeda dengan tampilan pada web, sehingga diperlukan sedikit modikasi untuk menyesuaikannya
# Kolom 10

for row in data:
    hasil_str = row[-1]  # ambil elemen terakhir (string hasil pertandingan)

    # 1 Ambil kata ke-2 sampai ke-5
    hasil_str = hasil_str.split()[1:6]
    hasil_str = ' '.join(hasil_str)

    # 2 Ganti pola teks sesuai instruksi
    hasil_str = (hasil_str
                 .replace('WinWResult', 'W')
                 .replace('LossLResult', 'L')
                 .replace('DrawDResult', 'D')
                 .replace('WinDResult', 'W')
                 .replace('WinLResult', 'W')
                 .replace('LossDResult', 'L')
                 .replace('LossWResult', 'L')
                 .replace('DrawLResult', 'D')
                 .replace('DrawWResult', 'D'))

    # 3 Simpan kembali ke data
    row[-1] = hasil_str

# Cek hasil
for d in data:
    print(d[-1])
```

```python
 # List data to DataFrame

df = pd.DataFrame(data, columns = headers)
df
```
